```python
#importing essential libraries
#learn more about phonenumbers https://www.abstractapi.com/guides/validate-phone-number-python
#essential library to validate phonenumbers
import phonenumbers
from datetime import datetime
from reportlab.platypus import SimpleDocTemplate, Table, Paragraph, TableStyle
from reportlab.lib import colors
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet


#creating circles dictionary to store circle id and their description
circles = {'KA':'Karnataka','AP':'Andhra Pradesh & Telangana','AS':'Assam','BR':'Bihar & Jharkhand','DL':'Delhi','GJ':'Gujrat',
          'HP':'Himachal Pradesh','HR':'Haryana','JK':'Jammu and Kashmir','KL':'Kerala and Lakshadweep','KO':'Kolkata',
          'MH':'Maharashtra and Goa','MP':'Madhya Pradesh and Chhattisgarh','MU':'Mumbai','NE':'North East','OR':'Orissa', 
          'PB':'Punjab','RJ':'Rajasthan','TN':'Tamil Nadu','UE':'UP(East)','UW':'UP(West)','WB':'West Bengal',
          'GB':'Ghaziabad and Noida'}


#creating prepaid packs 2d dictionary to store recharge types and description
prepaid_packs = {'full_recharge':['Rs.250:Unlimited calls (local/national) + 1.5GB data/day + 100 SMS/day. val = 28days',
                                   'Rs.400:Unlimited calls (local/national) + 3GB data/day + 100 SMS/day. val = 28days' ,
                                   'Rs.450:Unlimited calls (local/national) + 1.5GB data/day + 100 SMS/day. val = 56days' ,
                                    'Rs.500:Unlimited calls (local/national) + 1.5GB data/day + 100 SMS/day. val = 70days' ,
                                    'Rs.550:Unlimited calls (local/national) + 1.5GB data/day + 100 SMS/day. val =77days'],
                'Talktime_only':['Rs.100:Talktime: Rs.82 val = 28days',
                                'Rs.50:Talktime: Rs.39 val = 28days',
                                'Rs.30:Talktime: Rs.22 val = 28days',
                                'Rs.20:Talktime: Rs.14 val = 28days',
                                'Rs.10: Talktime: Rs.7 val = 28days'],
                'Data_Packs':['Rs.1200:240GB data val = 240days',
                             'Rs.600:72GB data val = 70days',
                             'Rs.250:50GB data val = 28days',
                             'Rs.100:12GB data val = 28days',
                             'Rs.50:6GB data val = 28days']}

#function for payment methods
def payment():
    while(1):
        print("please choose the payment methods \n")
        print("press 1 for Credit Card\n")
        print("press 2 for Debit Card \n")
        print("press 3 for UPI \n")
        print("press 4 for Net Banking \n")
        print("press 5 for Cash \n")
        
        ch = int(input())
        
        if(ch == 1):
            return "Credit card"
        elif(ch == 2):
            return "Debit card"
        elif(ch == 3):
            return "UPI"
        elif(ch == 4):
            return "Net Banking"
        elif(ch == 5):
            return "Cash"
        else:
            print("enter a valid choice")
        

    
#function to create reciept
#code is taken from geeksforgeeks website
def create_reciept(customer_name,phone,circle,choice_desc,prepaid_recharge_desc,recharge_desc,payment_method,total_bill):
  # data which we are going to display as tables
    DATA = [
           [ "Customer Name" , customer_name ],
           ["Phone Number",phone],
           [ "Circle ID",circle],
           [ "Date and Time of recharge",datetime.now()],
           [ "Recharge Type",choice_desc],
           [ "Prepaid Recharge Type",prepaid_recharge_desc],
           ["Plan selected ",recharge_desc],
           ["payment_method",payment_method],
           ["Total amount to be paid",total_bill]
           ]

   # creating a Base Document Template of page size A4
    pdf = SimpleDocTemplate( "receipt.pdf" , pagesize = A4 )
    styles = getSampleStyleSheet()
    title_style = styles[ "Heading1" ]
    title_style.alignment = 1
    title = Paragraph( "Recharge Summary" , title_style )
    style = TableStyle(
      [
         ( "BOX" , ( 0, 0 ), ( -1, -1 ), 1 , colors.black ),
         ( "GRID" , ( 0, 0 ), ( 4 , 4 ), 1 , colors.black ),
         ( "BACKGROUND" , ( 0, 0 ), ( 3, 0 ), colors.gray ),
         ( "TEXTCOLOR" , ( 0, 0 ), ( -1, 0 ), colors.whitesmoke ),
         ( "ALIGN" , ( 0, 0 ), ( -1, -1 ), "CENTER" ),
         ( "BACKGROUND" , ( 0 , 1 ) , ( -1 , -1 ), colors.beige ),
      ]
     )
    table = Table( DATA , style = style )
    pdf.build([ title , table ])
    


#main function to define all the procedure or logic    
def main():
    customer_name = input("enter the name of a customer : \n")
    
    #adding while loop to continously ask the customer until user gives right phone number
    while(1):
        print("enter the phonenumber in the format +countrycode followed by ten digit number \n")
        phone = input("enter the phone number of customer \n")
    
        #validating the phone number of customer using phonenumber
        phone = phonenumbers.parse(phone)
        if(phonenumbers.is_valid_number(phone)):
             break
        else:
            print("enter a valid phone number")
    
    
    while(1):
        circle_id = input("enter the circle id : \n")
        #validating the circle id entered by customer
        if(circle_id in circles.keys()):
            break
        else:
            print("enter valid circle_id")
    
    while(1):
       #giving users the option to choose among the prepaid and postpaid charges
        print("press 1 for prepaid recharge \n")
        print("press 2 for postpaid recharge \n")
    
        choice = int(input())
    
        if(choice == 1):
            choice_desc = 'Prepaid Recharge'
            while(1):
                print("press 1 for Full Recharge \n")
                print("press 2 for Talktime only \n")
                print("press 3 for Data packs \n ")
        
                recharge_pack_choice = int(input())
        
                if(recharge_pack_choice == 1):
                    prepaid_recharge_desc = "Full Recharge"
                    k = 1
                    for i in prepaid_packs['full_recharge']:
                        print("press {} for {} \n".format(k,i))
                        k = k+1
            
                    full_recharge_choice = int(input())
                    recharge_desc = prepaid_packs['full_recharge'][full_recharge_choice-1]
                    payment_method = payment()
                    #calculating the total bill amount of the recharge
                    total_bill = int(recharge_desc.split(".")[1].split(":")[0]) + (18/100)
                    print("Your Recharge Reciept has been Generated and downloaded Successfully..")
                    break
            
                elif(recharge_pack_choice == 2):
                    prepaid_recharge_desc = 'Talktime only'
                    k = 1
                    for i in prepaid_packs['Talktime_only']:
                        print("press {} for {} \n".format(k,i))
                        k = k+1
                    
                    full_recharge_choice = int(input())
                    recharge_desc = prepaid_packs['Talktime_only'][full_recharge_choice-1]
                    payment_method = payment()
                    #calculating the total bill amount of the recharge
                    total_bill = int(recharge_desc.split(".")[1].split(":")[0]) + (18/100)
                    print("Your Recharge Reciept has been Generated and downloaded Successfully..")
                    break
                
                elif(recharge_pack_choice == 3):
                    prepaid_recharge_desc = 'Data Packs'
                    k = 1
                    for i in prepaid_packs['Data_Packs']:
                        print("press {} for {} \n".format(k,i))
                        k = k+1
                    
                    full_recharge_choice = int(input())
                    recharge_desc = prepaid_packs['Data_Packs'][full_recharge_choice-1]
                    payment_method = payment()
                    #calculating the total bill amount of the recharge
                    total_bill = int(recharge_desc.split(".")[1].split(":")[0]) + (18/100)
                    print("Your Recharge Reciept has been Generated and downloaded Successfully..")
                    break
                
                else:
                    print("please enter a valid prepaid recharge choice \n")
            break
                    
        elif(choice == 2):
            choice_desc = 'Postpaid Recharge'
            print("Please enter the bill amount you want to recharge for \n")
            cost = int(input())
            cost_in = "Rs."+" "+str(cost)
            payment_method = payment()
            total_bill = cost + (18/100)
            print("Your Recharge Reciept has been Generated and downloaded Successfully..")
            break
            
        else:
            print("Enter a valid choice")
            
    if(choice == 1):
        create_reciept(customer_name,phone,circles[circle_id],choice_desc,prepaid_recharge_desc,recharge_desc,payment_method,total_bill)
    elif(choice == 2):
        create_reciept(customer_name,phone,circles[circle_id],choice_desc,"NA",cost_in,payment_method,total_bill)

            
if __name__== "__main__" :
    main()
            
    
        
```
